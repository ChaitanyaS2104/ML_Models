## Importing dependencies and dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('insurance.csv')
```

```python
df.head()
```

```python
# Features
X = df.drop(['charges'], axis=1)
# Labels
y = df['charges']
```

```python
X.head()
```

```python
y.head()
```

## Data preprocessing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler, PolynomialFeatures
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
# Categorical columns
columns_to_encode = ['sex', 'smoker','region']
# Numeric columns
numeric_cols = ['age','bmi','children']

numeric_pipeline = Pipeline([
    ('poly', PolynomialFeatures(degree=3, include_bias=False)),
    ('scaler', StandardScaler()),
])

preprocessor = ColumnTransformer([
    ('num', numeric_pipeline, numeric_cols),
    ('encoder', OneHotEncoder(sparse_output=False), columns_to_encode)
],
    remainder = 'passthrough'                             
)

transformed_train_data = preprocessor.fit_transform(X_train)
transformed_test_data = preprocessor.transform(X_test)

ohe = preprocessor.named_transformers_['encoder']
ohe_features = ohe.get_feature_names_out(columns_to_encode)

poly = preprocessor.named_transformers_['num']
poly_features = poly.get_feature_names_out(numeric_cols)
all_cols = list(poly_features) + list(ohe_features)

X_train = pd.DataFrame(transformed_train_data, columns=all_cols)
X_test = pd.DataFrame(transformed_test_data, columns=all_cols)
```

```python
X_test.head()
```

```python
X_train.head()
```

## Train the model

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
y_train_pred = lin_reg.predict(X_train)
y_test_pred = lin_reg.predict(X_test)
```

## Evaluate the model

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_test_pred)
```

## Plot

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=X_train['age'], y=y_train, color='purple', alpha=0.6)
plt.plot(X_train['age'], y_train_pred, color='orange')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Charges vs Age')
plt.show()
```
