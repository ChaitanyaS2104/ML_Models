## Import libraries and dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('insurance.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
# No missing values in the dataset
```

```python
# Features
X = df.drop(['charges'], axis=1)
# Labels
y = df['charges']
```

## Data preprocessing

```python
# Split the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Encoding categorical labels & apply scaling
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
# Categorical columns
columns_to_encode = ['sex', 'smoker','region']
# Numeric columns
numeric_cols = ['age','bmi','children']

preprocessor = ColumnTransformer(
    transformers=[
        ('scaler', StandardScaler(), numeric_cols),
        ('encoder', OneHotEncoder(sparse_output=False), columns_to_encode)
    ],
    remainder='passthrough'
)

transformed_data_train = preprocessor.fit_transform(X_train)
transformed_data_test = preprocessor.transform(X_test)

ohe = preprocessor.named_transformers_['encoder']
ohe_feature_names = ohe.get_feature_names_out(columns_to_encode)
all_columns = numeric_cols + list(ohe_feature_names)

X_train = pd.DataFrame(transformed_data_train, columns=all_columns)
X_test = pd.DataFrame(transformed_data_test, columns=all_columns)
```

## Train model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
reg = RandomForestRegressor(n_estimators=8, min_samples_leaf=8)
reg = reg.fit(X_train, y_train)
print(f'Score on train data: {reg.score(X_train, y_train)}')
y_train_pred = reg.predict(X_train)
```

## Evaluate model

```python
from sklearn.metrics import r2_score
```

```python
y_test_pred = reg.predict(X_test)
r2_score(y_test, y_test_pred)
```

## Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,6))
# Train data
plt.scatter(y_train, y_train_pred, alpha=0.6, color='purple')
# Test data
plt.scatter(y_test, y_test_pred, alpha=0.6, color='orange')
plt.xlabel('Actual price')
plt.ylabel('Predicted price')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'k--', lw=2)
plt.title('Actual vs Predicted')
plt.show()
```

```python
# Train data
plt.figure(figsize=(4,4))
plt.scatter(y_train, X_train['age'], alpha=0.8, color='purple')
plt.plot(y_train_pred, X_train['age'], color='orange')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Charges vs Age')
plt.show()
```

```python
# Test data
plt.figure(figsize=(4,4))
plt.scatter(y_test, X_test['age'], alpha=0.8, color='purple')
plt.plot(y_test_pred, X_test['age'], color='orange')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Charges vs Age')
plt.show()
```
